```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import DeepFace
```

```python
cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
cascade_smile=cv2.CascadeClassifier('haarcascade_smile.xml')
cascade_eye=cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
image=cv2.imread('lena.JPG')
rgb=cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
print(rgb.shape)
print(rgb.ndim)
plt.imshow(rgb)
```

```python
face=cascade.detectMultiScale(rgb,1.1,3)
print(face)
```

```python
#x=218,y=202,w=169,h=169
```

```python
plt.imshow(rgb[218:218+169,202:202+169])
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,255,0),3)
    roi=rgb[y:y+h,x:x+w]
    smile=cascade_smile.detectMultiScale(roi,1.1,15)
    for (x,y,w,h) in smile:
        cv2.rectangle(roi,(x,y),(x+w,y+h),(0,0,255),3)
    eye =cascade_eye.detectMultiScale(roi,1.1,3)
    for(x,y,w,h) in eye:
        cv2.rectangle(roi,(x,y),(x+w,y+h),(255,0,0),3)
        
plt.imshow(rgb)    
```
